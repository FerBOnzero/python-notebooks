## <span style='color:Red'> Ejercicio III </span>:
#### Para todas las variables categóricas, obtén la frecuencia, frecuencia relativa y frecuencia acumulada y describe cuál es la moda o modas de cada variable.

### Definimos la variable "indices" para sólo acceder a los valores categóricos

```python
indices = list(df_resumen[df_resumen['Q1'].isna()].index)
indices
```

### Hagamos la lista de frecuencias absolutas para que apartir de aquí podamos calcular la frecuencia relativa

```python
lista_frecuencias_absolutas = []
for i in indices:
    empty_list = []
    empty_list.append(pd.value_counts(df_cars93[i]).to_list())
    lista_frecuencias_absolutas.append(empty_list)
```

### Haremos uso de un diccionario para poder convertirlo a DataFrame conservando el órden de las variables de df_cars93

```python
dict_init = {}
count = 0
for i in indices:
    dict_init[i] = lista_frecuencias_absolutas[count]
    count += 1
```

### A partir de aquí definamos "tabla_frecuencias" para poder almacenar en un DataFrame las frecuencias que vayamos calculando

```python
tabla_frecuencias = pd.DataFrame(dict_init).T
tabla_frecuencias.set_axis(['Frecuencia absoluta'],  axis='columns', inplace = True)
tabla_frecuencias
```

### Calculemos la frecuencia relativa haciendo uso de la fórmula y almacenar los valores en una lista llamada "lista_frecuencias_relativas"

```python
count = 0
lista_frecuencias_relativas = []

for i in tabla_frecuencias['Frecuencia absoluta']:
    empty_list = []
    for j in i:
        empty_list.append(j/len(df_cars93[indices[count]]))
    lista_frecuencias_relativas.append([empty_list])
    count += 1
```

### Al igual que en frecuencias absolutas, almacenaremos los valores en un diccionario para así poder hacerlo DataFrame consevando su orden y poder "concatenar" ambos DataFrames

```python
dict_init_relativo = {}
count = 0
for i in indices:
    dict_init[i] = lista_frecuencias_relativas[count]
    count += 1
```

```python
tabla_frecuencias_relativas = pd.DataFrame(dict_init).T
tabla_frecuencias_relativas.set_axis(['Frecuencia relativa'],  axis='columns', inplace = True)

tabla_frecuencias_relativas
```

### Unamos los dos DataFrame hechos hasta ahora

```python
tabla_frecuencias = pd.concat([tabla_frecuencias, tabla_frecuencias_relativas], axis=1)
tabla_frecuencias
```

### Comprobemos dos cosas aquí: 
- El número de elementos en cada frecuencia absoluta de las variables es la misma que el número de elementos de frecuencia relativa
- La suma de cada conjunto de datos de frecuencia relativa es 1

```python
proof_relativa = []
proof_absoluta = []
for i in tabla_frecuencias['Frecuencia relativa']:
    proof_relativa.append(len(i))
for i in tabla_frecuencias['Frecuencia absoluta']:
    proof_absoluta.append(len(i))
proof_absoluta == proof_relativa
```

```python
proof_sum_relativa = []
for i in tabla_frecuencias['Frecuencia relativa']:
    proof_sum_relativa.append(sum(i))
proof_sum_relativa
```

### Calculemos la frecuencia acumulada. 


```python
def frec_acumulativa(n):
    if len(n) == 1:
        return n
    else:
        a=[]
        for i in range(1,len(n)+1):
            a.append(sum(n[:i]))
        return a
```

### Hagamos la lista de frecuencias acumulativas

```python
count = 0
lista_frecuencias_acumulativas = []

for i in tabla_frecuencias['Frecuencia relativa']:
    lista_frecuencias_acumulativas.append([frec_acumulativa(i)])
```

### Almacenaremos los valores en un diccionario para así poder hacerlo DataFrame consevando su orden y poder "concatenar" al DataFrame que hemos estado trabajando

```python
dict_init_acumulativo = {}
count = 0
for i in indices:
    dict_init_acumulativo[i] = lista_frecuencias_acumulativas[count]
    count += 1
```

```python
tabla_frecuencias_acumulativas = pd.DataFrame(dict_init_acumulativo).T
tabla_frecuencias_acumulativas.set_axis(['Frecuencia acumulativa'],  axis='columns', inplace = True)

tabla_frecuencias_acumulativas
```

### Unamos los DataFrame hechos hasta ahora

```python
tabla_frecuencias = pd.concat([tabla_frecuencias, tabla_frecuencias_acumulativas], axis=1)
tabla_frecuencias
```

### Calculemos ahora la moda, en la cual nos apoyaremos de la función "mode"

```python
def mode(dataset):
    frequency = {}

    for value in dataset:
        frequency[value] = frequency.get(value, 0) + 1

    most_frequent = max(frequency.values())

    modes = [key for key, value in frequency.items()
                      if value == most_frequent]

    return modes
```

### Hagamos la lista de moda

```python
count = 0
moda_lista = []

for i in indices:
    moda_lista.append([mode(df_cars93[i])])
```

### Almacenaremos los valores en un diccionario para así poder hacerlo DataFrame consevando su orden y poder "concatenar" al DataFrame que hemos estado trabajando

```python
dict_init_moda = {}
count = 0
for i in indices:
    dict_init_moda[i] = moda_lista[count]
    count += 1
```

```python
tabla_moda = pd.DataFrame(dict_init_moda).T
tabla_moda.set_axis(['Moda'],  axis='columns', inplace = True)

tabla_moda
```

### Unamos los DataFrame hechos hasta ahora, finalizando así la tabla frecuencia

```python
tabla_frecuencias = pd.concat([tabla_frecuencias, tabla_moda], axis=1)
tabla_frecuencias
```

### Generemos las gráficas de dispersión de nuestras variables (sin contemplar variables iguales)

```python
def graficos_dispersion(df1, df2, i, j):
    fig, ax = plot.subplots(1, 1, figsize=(6,4))
    ax.scatter(x=df1, y=df2, alpha= 0.8)
    ax.set_xlabel(i)
    ax.set_ylabel(j);
```

```python
list_correlacion = ['Fuel.tank.capacity', 'Wheelbase', 'Rear.seat.room', 'Width', 'Length']
for i in list_correlacion:
    list_copy = list_correlacion.copy()
    list_copy.pop(list_copy.index(i))
    for j in list_copy: 
        graficos_dispersion(df_cars93[i], df_cars93[j], i, j)
    list_correlacion.pop(list_correlacion.index(i))
```
