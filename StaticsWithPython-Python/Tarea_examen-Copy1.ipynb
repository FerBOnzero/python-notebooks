## Ejercicio II

```python
pd.read_csv('df_Resumen.csv')
```

```python
df_resumen = df_cars93.describe(include='all').T
df_resumen = df_resumen[['mean', 'std', 'min', 'max', '25%', '75%']]
```

```python
kurt = df_cars93.kurt(axis=0, numeric_only=True)
kurt = pd.DataFrame(kurt)
kurt.set_axis(['Kurtosis'], axis='columns', inplace=True)
```

```python
rango_muestral = pd.DataFrame(df_resumen['max'] - df_resumen['min'])
rango_muestral.set_axis(['Rango muestral'], axis='columns', inplace=True)
```

```python
mediana = pd.DataFrame(df_cars93.median())
mediana.set_axis(['Mediana'], axis='columns', inplace=True)
```

```python
sesgo = pd.DataFrame(df_cars93.skew(axis = 0, skipna = True))
sesgo.set_axis(['Sesgo'], axis='columns', inplace=True)
```

```python
total_nan = pd.DataFrame(df_cars93.isna().sum())
total_nan.set_axis(['Valores perdidos'], axis='columns', inplace=True)
```

```python
df_resumen = pd.concat([df_resumen, kurt, rango_muestral, mediana, sesgo, total_nan], axis=1)
df_resumen
```

```python
df_resumen.set_axis(['Promedio', 'Desviación estándar', 'Mínimo', 'Máximo', 'Q1', 'Q3', 'Kurtosis', 
                     'Rango muestral', 'Mediana', 'Sesgo', 'Valores perdidos'], axis='columns', inplace=True)
df_resumen
```
